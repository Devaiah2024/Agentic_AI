```python
import yaml
from crewai import LLM,Agent,Task,Crew
```

```python
llm=LLM(
    model="ollama/llama3.2",
    base_url="http://localhost:11434",
    api_key="ollama"
)
```

```python
llm
```

```python
with open("config_data.yaml","r") as file:
    config=yaml.safe_load(file)
```

```python
config
```

```python
config["agents"]["role"]
```

```python
simple_agent=Agent(
    role=config["agents"]["role"],
    goal=config["agents"]["goal"],
    backstory=config["agents"]["backstory"],
    llm=llm,
    verbose=True
)
```

```python
simple_agent
```

```python
simple_task=Task(
    description=config["tasks"]["description"],
    agent=simple_agent,
    expected_output=config["tasks"]["expected_output"]

)
```

```python
simple_task
```

```python
crew=Crew(
    agents=[simple_agent],
    tasks=[simple_task],
    verbose=True
)
```

```python
reposnse=crew.kickoff()
```

```python
print(reposnse)
```

# Multiple agets ans tasks

```python
with open("config2.yaml","r") as files:
    config=yaml.safe_load(files)
```

```python
config
```

```python
researcher_agent1=Agent(
    role=config["agents"]["researcher"]["role"],
    goal=config["agents"]["researcher"]["goal"],
    backstory=config["agents"]["researcher"]["backstory"],
    allow_delegation=config["agents"]["researcher"]["allow_delegation"],
    llm=llm,
    verbose=True
)
```

```python
writer_agent=Agent(
    role=config["agents"]["writer"]["role"],
    goal=config["agents"]["writer"]["goal"],
    backstory=config["agents"]["writer"]["backstory"],
    allow_delegation=config["agents"]["writer"]["allow_delegation"],
    llm=llm,
    verbose=True
)
```

```python

researcher_task=Task(
    description=config["tasks"]["research_task"]["description"],
    agent=researcher_agent1,
    expected_output=config["tasks"]["research_task"]["expected_output"]
)
```

```python
wtriter_task=Task(
    description=config["tasks"]["writing_task"]["description"],
    agent=writer_agent,
    expected_output=config["tasks"]["writing_task"]["expected_output"]
    
)
```

```python
crew=Crew(
    agents=[researcher_agent1,writer_agent],
    tasks=[researcher_task,wtriter_task],
    verbose=True
)
```

```python
response=crew.kickoff()
```

```python
print(response)
```
