```python
from crewai import LLM,Agent,Task,Crew
llm=LLM(
    model="ollama/llama3.2",
    base_url="http://localhost:11434",
    api_key="ollama"
)

simple_agent=Agent(
    role="Generative AI and LLM model developer",
    goal="Explain about Agentic AI from start to end with code",
    backstory="Summarize a detailed roadmap to become an Agentic AI developer and how to get a high-paying job.",
    llm=llm,
    verbose=True
)

simple_task = Task(
    description="""
    Summarize a detailed roadmap to become an Agentic AI developer
    and how to land a high-paying job. Highlight the biggest AI advancements.
    """,
    agent=simple_agent,
    expected_output="A full analysis report on Agentic AI jobs and advancements in 2026."
)

crew=Crew(
    agents=[simple_agent],
    tasks=[simple_task],
    verbose=True
)
response=crew.kickoff()
print(response)




```
